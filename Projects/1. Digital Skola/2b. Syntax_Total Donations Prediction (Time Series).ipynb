```python
from sklearn.preprocessing import MinMaxScaler
from keras.preprocessing.sequence import TimeseriesGenerator
from pandas.tseries.offsets import DateOffset
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout

#Performance 
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np
```

```python
train_donasi_guest = donasi_guest1[:len(donasi_guest1)-30]
test_donasi_guest = donasi_guest1[len(donasi_guest1)-30:]
```

```python
train_donasi_guest.head()
```

```python
scaler = MinMaxScaler()
scaler.fit(train_donasi_guest)
train = scaler.transform(train_donasi_guest)
test = scaler.transform(test_donasi_guest)
```

```python
n_input = 30
n_features = 1
generator = TimeseriesGenerator(train, train, length=n_input, batch_size=6)
```

```python
model = Sequential()
model.add(LSTM(200, activation='relu', input_shape=(n_input, n_features)))
model.add(Dropout(0.15))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
model.fit_generator(generator,epochs=90, verbose= 0)
model.summary()
```

```python
pred_list = []
batch = train[-n_input:].reshape((1, n_input, n_features))
for i in range(n_input):   
    pred_list.append(model.predict(batch)[0]) 
    batch = np.append(batch[:,1:,:],[[pred_list[i]]],axis=1)
```

```python
df_predict= pd.DataFrame(scaler.inverse_transform(pred_list),index=donasi_guest1[-n_input:].index, columns=['Prediction'])
```

```python
df_predict[:len(donasi_guest1)-30]=np.nan
df_predict= pd.DataFrame(scaler.inverse_transform(pred_list),index=donasi_guest1[-n_input:].index, columns=['Prediction'])
df_test = pd.concat([donasi_guest1,df_predict], axis=1)
```

```python
def forecasting(data):
    test = data['new_total'].iloc[len(data)-round(len(data)/5):]
    train = data['new_total'].iloc[:len(data)-round(len(data)/5)]
    # Build Model
    model = ARIMA(train, order = data['orde'][0])
    fitted = model.fit(disp = -1)
    print(fitted.summary())

    # Forecast
    global fc
    fc, se, conf = fitted.forecast(round(len(data)/5)+30, alpha =0.05) #95% CI
    # Make as pandas series
    fc_series = pd.Series(fc, index = test.index)
    lower_series = pd.Series(conf[:, 0], index = test.index)
    upper_series = pd.Series(conf[:, 1], index = test.index)

    # Plot 
    plt.figure(figsize=(12,5), dpi = 100)
    plt.plot(train, label = 'training')
    plt.plot(test, label = 'actual')
    plt.plot(fc_series, label = 'forecast')
    plt.fill_between(lower_series.index, lower_series, upper_series, 
                     color = 'k', alpha = 0.2)
    plt.title('Forecast VS Actuals')
    plt.legend(loc = 'upper left', fontsize = 8)
    plt.show()

    # Accuracy metrics
    mape = np.mean(np.abs(fc - test)/np.abs(test))  # MAPE
    me = np.mean(fc - test)             # ME
    mae = np.mean(np.abs(fc - test))    # MAE
    mpe = np.mean((fc - test)/test)   # MPE
    rmse = np.mean((fc - test)**2)**.5  # RMSE
    mins = np.amin(np.hstack([fc[:,None], 
                              test[:,None]]), axis=1)
    maxs = np.amax(np.hstack([fc[:,None], 
                              test[:,None]]), axis=1)
    minmax = 1 - np.mean(mins/maxs)             # minmax
    print('mape : {}\nme: {}\nmae : {}\nmpe : {}\nrmse : {}\nminmax : {}'.format(mape,me, mae,mpe,rmse,minmax))
    forecasting.fc = fc
```
